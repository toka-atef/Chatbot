```python
df=pd.read_csv('PURCHASE ORDER DATA EXTRACT 2012-2015_0.csv')
pd.set_option('display.max_columns', None)

df.info()
```

```python
df['Normalized UNSPSC'].value_counts(dropna=False)
```

```python
df['Creation Date'] = pd.to_datetime(df['Creation Date'])
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'],errors='coerce')
```

```python
df['Supplier Code']=df['Supplier Code'].fillna(0).astype(int)
df['Quantity']=df['Quantity'].fillna(0).astype(int)
df['Segment']=df['Segment'].fillna(0).astype(int)
df['Family']=df['Family'].fillna(0).astype(int)
df['Class']=df['Class'].fillna(0).astype(int)
df['Normalized UNSPSC']=df['Normalized UNSPSC'].fillna(0).astype(int)
df.info()
```

```python
df.columns
```

```python
df['Fiscal Year']=df['Fiscal Year'].fillna(0)
```

```python
df['Creation Month']=df['Creation Date'].dt.month
df['Creation year']=df['Creation Date'].dt.year
```

```python
def get_Fiscal_Year():
    for i in range(len(df)):
        if pd.isna(df["Fiscal Year"].iloc[i]):
            if df['Creation Month']>6:
                string=str(df['Creation year'].iloc[i])+"-"+str((df['Creation year'].iloc[i])+1)
            elif df['Creation Month']<=6:
                string=str((df['Creation year'].iloc[i])-1)+"-"+str(df['Creation year'].iloc[i])
            df["Fiscal Year"].iloc[i]=string   
get_Fiscal_Year() 
```

Merge 2 datasets together 

```python
merged1 = pd.merge(df, type, on='Acquisition Method', how='left')

unmatched = merged1[merged1['Acquisition Method Description'].isna()]
merged2 = pd.merge(unmatched.drop(columns=['Acquisition Method Description']), 
                   type, left_on='Acquisition Method', right_on='Abbreviation', how='left')

final_result = pd.concat([merged1[~merged1['Acquisition Method Description'].isna()], merged2])
final_result = final_result.drop(columns=['Abbreviation'])

final_result
```

```python
final_result
```

```python
final_result['Acquisition Method'].fillna(final_result['Acquisition Method_y'],inplace=True)
final_result['Acquisition Method'].fillna(final_result['Acquisition Method_x'],inplace=True)
```

```python
final_result["Acquisition Method"].value_counts(dropna=True)
```

```python
final_result[["Acquisition Method",'Acquisition Method_x'	,'Abbreviation_x'	,'Acquisition Method_y'	,'Abbreviation_y']][final_result['Abbreviation_y'].isna()].isna().sum()
```

```python
final_result.drop(["Acquisition Method_x","Abbreviation_y","Acquisition Method_y"],axis=1,inplace=True)
```

```python
final_result.rename(columns={"Abbreviation_x":"Abbreviation"},inplace=True)
final_result['Abbreviation'].fillna("No Abbreviation",inplace=True)
```

```python
final_result
```

```python
final_result["Location"].fillna("Not Spacified",inplace=True)
```

```python
final_result['final_result Price'] = final_result['Unit Price'].str.replace("$", "").astype(float)
final_result['Total Price']=final_result['Unit Price']*final_result['Quantity']
```

```python
final_result[final_result['Acquisition Method Description'].isna()]['Acquisition Method'].value_counts()
```

```python
final_result=final_result.drop("Purchase Date",axis=1)
```

```python
# save the merged csv 
final_result.to_csv('CA purchase.csv',index=False)
```
